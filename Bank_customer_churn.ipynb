```python
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OrdinalEncoder

def preprocessCustomersPipeline(data):
    """Prepares the original customers data table in a form ready for the Deep Learning model, using
    all the transformations and edits we used above, and returns a ready dataset"""

    data = data.drop('Naive_Bayes_Classifier_Attrition_Flag_Card_Category_Contacts_Count_12_mon_Dependent_count_Education_Level_Months_Inactive_12_mon_1',1)
    data = data.drop('Naive_Bayes_Classifier_Attrition_Flag_Card_Category_Contacts_Count_12_mon_Dependent_count_Education_Level_Months_Inactive_12_mon_2', 1)
    data = data.drop('CLIENTNUM', 1)
    
    #Categorical Features
    a_map = {'Attrited Customer':1, 'Existing Customer':0}
    data['Attrition_Flag'] = data['Attrition_Flag'].map(a_map)
    g_map = {'M':0, 'F':1}
    data['Gender'] = data['Gender'].map(g_map)
    
    marital = OrdinalEncoder(categories=[['Unknown','Single','Married','Divorced']])
    data['Marital_Status'] = marital.fit_transform(data[['Marital_Status']])

    edu = OrdinalEncoder(categories=[['Unknown', 'Uneducated', 'High School', 'College','Graduate',
                                      'Post-Graduate','Doctorate']])
    data['Education_Level'] = edu.fit_transform(data[['Education_Level']])

    income = OrdinalEncoder(categories=[['Unknown','Less than $40K','$40K - $60K','$60K - $80K',
                                         '$80K - $120K','$120K +']])
    data['Income_Category'] = income.fit_transform(data[['Income_Category']])

    card = OrdinalEncoder(categories=[['Blue', 'Silver','Gold','Platinum']])
    data['Card_Category'] = card.fit_transform(data[['Card_Category']])
    
    #Numerical Features
    credit = StandardScaler()
    data['Credit_Limit'] = credit.fit_transform(data[['Credit_Limit']])
    revolv_bal = StandardScaler()
    data['Total_Revolving_Bal'] = revolv_bal.fit_transform(data[['Total_Revolving_Bal']])
    avg_open = StandardScaler()
    data['Avg_Open_To_Buy'] = avg_open.fit_transform(data[['Avg_Open_To_Buy']])
    trans = StandardScaler()
    data['Total_Trans_Amt'] = trans.fit_transform(data[['Total_Trans_Amt']])
    
    nums = ['Credit_Limit','Total_Revolving_Bal','Avg_Open_To_Buy','Total_Trans_Amt', 'Total_Trans_Ct']
    for n in nums:
        s = StandardScaler()
        data[n] = s.fit_transform(data[[n]])
    return data
```

**üìÅ 4. Import Library & Utility Function**
===========================================

Mengimpor semua pustaka yang diperlukan untuk preprocessing, resampling, modeling, dan evaluasi. Semua fungsi di atas bertugas untuk visualisasi dan evaluasi model dipanggil di bagian akhir nanti saat model sudah dilatih dan diuji.



```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from collections import Counter
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import SMOTE
from imblearn.under_sampling import TomekLinks
from imblearn.combine import SMOTETomek
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.metrics import classification_report, confusion_matrix, roc_curve, auc
from sklearn.metrics import precision_recall_curve, average_precision_score
from sklearn.model_selection import train_test_split

# Fungsi Confusion Matrix
def plot_confusion_matrix(cm, title):
    plt.figure(figsize=(6, 4))
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=["Class 0", "Class 1"], yticklabels=["Class 0", "Class 1"])
    plt.xlabel("Predicted Label")
    plt.ylabel("True Label")
    plt.title(title)
    plt.show()

# Fungsi Class Distribution
def plot_class_distribution(y, title):
    plt.figure(figsize=(6, 4))
    sns.countplot(x=y, palette="viridis")
    plt.title(title)
    plt.xlabel("Class")
    plt.ylabel("Count")
    plt.show()

# Fungsi Print Class Distribution
def print_class_distribution(y, dataset_name):
    counter = Counter(y)
    print(f"Distribusi kelas pada {dataset_name}: {counter}")

# ROC Curve

def plot_roc_pr_curves(y_true, y_prob, model_name, resample_method):
    fpr, tpr, _ = roc_curve(y_true, y_prob)
    roc_auc = auc(fpr, tpr)

    precision, recall, _ = precision_recall_curve(y_true, y_prob)
    pr_auc = average_precision_score(y_true, y_prob)

    fig, axs = plt.subplots(1, 2, figsize=(12, 5))

    axs[0].plot(fpr, tpr, label=f"AUC = {roc_auc:.4f}", color="blue")
    axs[0].plot([0, 1], [0, 1], linestyle="--", color="gray")
    axs[0].set_title(f"ROC Curve - {model_name} ({resample_method})")
    axs[0].set_xlabel("False Positive Rate")
    axs[0].set_ylabel("True Positive Rate")
    axs[0].legend(loc="lower right")
    axs[0].grid(True)

    axs[1].plot(recall, precision, label=f"AP = {pr_auc:.4f}", color="purple")
    axs[1].set_title(f"PR Curve - {model_name} ({resample_method})")
    axs[1].set_xlabel("Recall")
    axs[1].set_ylabel("Precision")
    axs[1].legend(loc="upper right")
    axs[1].grid(True)

    plt.tight_layout()
    plt.show()
```

=====================================
**‚úÇÔ∏è 5. Load Dataset dan Split Data**
=====================================

Data dibagi menjadi data latih dan data uji. Distribusi kelas divisualisasikan untuk melihat ketidakseimbangan.

```python
# Generate x and y sets
x = data.drop('Attrition_Flag', axis=1).values
y = data['Attrition_Flag']
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42, stratify=y)

# Distribusi sebelum resampling
print_class_distribution(y_train, "Training Set (Sebelum Resampling)")
print_class_distribution(y_test, "Testing Set (Sebelum Resampling)")
plot_class_distribution(y_train, "Original Training Class Distribution")
plot_class_distribution(y_test, "Original Testing Class Distribution")
```

==============================================
**‚åõ 6. Resampling (SMOTE, Tomek, SMOTE+Tomek)**
===============================================

Di sini tiap metode resampling dilakukan satu per satu dan ditampilkan hasil distribusinya.



```python
# SMOTE
smote = SMOTE(random_state=42)
X_smote, y_smote = smote.fit_resample(X_train, y_train)
print_class_distribution(y_smote, "Training Set (Setelah SMOTE)")
plot_class_distribution(y_smote, "SMOTE Class Distribution")

# Tomek Links
tomek = TomekLinks()
X_tomek, y_tomek = tomek.fit_resample(X_train, y_train)
print_class_distribution(y_tomek, "Training Set (Setelah Tomek Links)")
plot_class_distribution(y_tomek, "Tomek Links Class Distribution")

# SMOTE + Tomek
smote_tomek = SMOTETomek(random_state=42)
X_smote_tomek, y_smote_tomek = smote_tomek.fit_resample(X_train, y_train)
print_class_distribution(y_smote_tomek, "Training Set (Setelah SMOTE + Tomek)")
plot_class_distribution(y_smote_tomek, "SMOTE + Tomek Class Distribution")
```

====================================
**ü§Ø 7. Model Training dan Evaluasi**
====================================

Di sini bagian akhir yang melakukan loop terhadap semua kombinasi resampling dan model. Jadi kamu bisa lihat performa tiap skenario.



```python
print_class_distribution(y_test, "Testing Set (Setelah Resampling) [Sebenarnya Tidak Berubah]")
plot_class_distribution(y_test, "Testing Class Distribution (Tetap Sama)")

# Resampling methods
resampling_methods = {
    "Original": (X_train, y_train),
    "SMOTE": (X_smote, y_smote),
    "Tomek Links": (X_tomek, y_tomek),
    "SMOTE + Tomek": (X_smote_tomek, y_smote_tomek)
}

# Models
models = {
    "Random Forest": RandomForestClassifier(random_state=42),
    "XGBoost": XGBClassifier(use_label_encoder=False, eval_metric='logloss')
}

# Training, Evaluasi dan ROC
for resample_name, (X_resampled, y_resampled) in resampling_methods.items():
    print(f"\n=== Resampling Method: {resample_name} ===")
    for model_name, model in models.items():
        print(f"\nModel: {model_name}")
        
        # Train
        model.fit(X_resampled, y_resampled)
        
        # Predict
        y_pred = model.predict(X_test)
        y_prob = model.predict_proba(X_test)[:, 1]  # Probabilitas kelas 1
        
        # Evaluation
        print("Classification Report:")
        print(classification_report(y_test, y_pred, digits=4))
        
        print("Confusion Matrix:")
        cm = confusion_matrix(y_test, y_pred)
        print(cm)
        plot_confusion_matrix(cm, f"{model_name} - {resample_name}")
        
        # ROC Curve
        plot_roc_curve(y_test, y_prob, model_name, resample_name)
```

```python
def plot_roc_pr_curves(y_test, y_prob, model_name, resample_method):
    from sklearn.metrics import roc_curve, auc, precision_recall_curve, average_precision_score
    import matplotlib.pyplot as plt

    fpr, tpr, _ = roc_curve(y_test, y_prob)
    precision, recall, _ = precision_recall_curve(y_test, y_prob)
    roc_auc = auc(fpr, tpr)
    pr_auc = average_precision_score(y_test, y_prob)

    plt.figure(figsize=(12, 5))

    plt.subplot(1, 2, 1)
    plt.plot(fpr, tpr, label=f"AUC = {roc_auc:.4f}")
    plt.plot([0, 1], [0, 1], 'k--')
    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
    plt.title(f"ROC Curve - {model_name} - {resample_method}")
    plt.legend(loc="lower right")

    plt.subplot(1, 2, 2)
    plt.plot(recall, precision, label=f"AP = {pr_auc:.4f}")
    plt.xlabel("Recall")
    plt.ylabel("Precision")
    plt.title(f"PR Curve - {model_name} - {resample_method}")
    plt.legend(loc="lower left")

    plt.tight_layout()
    plt.show()
```
